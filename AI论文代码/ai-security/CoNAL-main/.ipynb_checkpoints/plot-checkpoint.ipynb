# 训练过程的折线图

````python
import numpy as np
import matplotlib.pyplot as plt

# 加载训练集、验证集准确率.npy文件
train_acc_self = np.load('./trainProcess/train_acc_self.npy')
val_acc_self = np.load('./trainProcess/valid_acc_self.npy')
train_acc_conal = np.load('./trainProcess/train_acc_conal.npy')
val_acc_conal = np.load('./trainProcess/valid_acc_conal.npy')
train_acc_xiaorong = np.load('./trainProcess/train_acc_xiaorong.npy')
val_acc_xiaorong = np.load('./trainProcess/valid_acc_xiaorong.npy')
````

````python
# 绘制self model
# 找到验证集准确率最大时的样本点
max_val_acc_index = np.argmax(val_acc_self)
max_val_acc = val_acc_self[max_val_acc_index]
max_val_acc_iteration = max_val_acc_index + 1  # 注意迭代次数从1开始


# 绘制折线图
iterations = np.arange(1, len(train_acc_self) + 1)
plt.plot(iterations, train_acc_self, label='Train Accuracy')
plt.plot(iterations, val_acc_self, label='Validation Accuracy')
plt.scatter(max_val_acc_iteration, max_val_acc, color='red', label=f'Max Val Acc: {max_val_acc:.2f},Epoch={(max_val_acc_index+1):d}', zorder=5)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Self Model Training and Validation Accuracy')
plt.legend()
plt.grid(True)
# 显示图形
plt.show()
````

````python
#绘制conal Model 训练过程
max_val_acc_index = np.argmax(val_acc_conal)
max_val_acc = val_acc_conal[max_val_acc_index]
max_val_acc_iteration = max_val_acc_index + 1  # 注意迭代次数从1开始
# 绘制折线图
iterations = np.arange(1, len(train_acc_conal) + 1)
plt.plot(iterations, train_acc_conal, label='Train Accuracy')
plt.plot(iterations, val_acc_conal, label='Validation Accuracy')
plt.scatter(max_val_acc_iteration, max_val_acc, color='red', label=f'Max Val Acc: {max_val_acc:.2f},Epoch={(max_val_acc_index+1):d}', zorder=5)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('CoNal Model Training and Validation Accuracy')
plt.legend()
plt.grid(True)
# 显示图形
plt.show()
````

````python
#绘制Abalation Study Model 训练过程
max_val_acc_index = np.argmax(val_acc_xiaorong)
max_val_acc = val_acc_xiaorong[max_val_acc_index]
max_val_acc_iteration = max_val_acc_index + 1  # 注意迭代次数从1开始
# 绘制折线图
iterations = np.arange(1, len(train_acc_xiaorong) + 1)
plt.plot(iterations, train_acc_xiaorong, label='Train Accuracy')
plt.plot(iterations, val_acc_xiaorong, label='Validation Accuracy')
plt.scatter(max_val_acc_iteration, max_val_acc, color='red', label=f'Max Val Acc: {max_val_acc:.2f},Epoch={(max_val_acc_index+1):d}', zorder=5)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Ablation Study Model Training and Validation Accuracy')
plt.legend()
plt.grid(True)
# 显示图形
plt.show()
````

# 绘制测试集的结果分析

````python
# 绘制不同模型在测试集上的准确率
import matplotlib.pyplot as plt

# 已知三个模型的测试准确率
model_names = ['Self Model', 'CoNal Model', 'Ablation Study Model',"Max-Mig","MW",
               "Majority Voting","Doctor Net",'AggNet',"crowdLayer","DL-DS","DL-MV","DL_MW"]
#加载准确率
self_acc=np.load("./testResult/test_acc.npy")
conal_acc=np.load("./testResult/test_acc_conal.npy")
xiaorong_acc=np.load("./testResult/test_acc_xiaorong.npy")
Max_Mig=np.load("./trainProcess/testResult/Max-MIG Acc.npy")
MV=np.load("./trainProcess/testResult/Majority Voting Acc.npy")
doctorNet=np.load("./trainProcess/testResult/DN Cotraining Acc.npy")
MW=np.load("./trainProcess/testResult/MW Cotraining Acc.npy")
AggNet=np.load("./trainProcess/testResult/AGG Cotraining Acc.npy")
# MBEM=np.load("./trainProcess/testResult/MBEM Cotraining Acc.npy")
crowdLayer=np.load("./trainProcess/testResult/crowdlayer.npy")
DL_DS=np.load("./trainProcess/testResult/DL-DS.npy")
DL_MV=np.load("./trainProcess/testResult/DL-MV.npy")
DL_MW=np.load("./trainProcess/testResult/DL-MW.npy")

test_accuracy = [self_acc, conal_acc, xiaorong_acc,Max_Mig,MW,
                 MV,doctorNet,AggNet,crowdLayer,DL_DS,DL_MV,DL_MW]
test_accuracy,xiaorong_acc,model_names
````

````python
# 绘制柱状图

plt.barh(model_names, test_accuracy, color='skyblue')
# 获取当前轴对象
ax = plt.gca()

# 设置纵坐标刻度和标签
ax.set_yticks(np.arange(len(model_names)))
ax.set_yticklabels(model_names)
# 在柱状图上显示小数点后一位的准确度
# 在柱状图上显示小数点后一位的准确度，并确保文字不与画布重合
for i in range(len(model_names)):
    plt.text(test_accuracy[i], i, f'{test_accuracy[i]:.1%}', ha='left', va='center')
# 强制设置横坐标的最大数值为 0.95
plt.xlim(right=1)
# 添加标题和标签
plt.xlabel('Test Accuracy')
plt.ylabel('Model')
plt.title('Test Accuracy of Different Models')
# 显示图形
plt.show()
````

# 单独显示
```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
model_names = ['Self Model', 'CoNal Model', 'Ablation Study Model']
categories = ["highway", "insidecity", "tallbuilding", "street", "forest", "coast", "mountain",
                           "opencountry"]
accuracy = np.load("./testResult/acc_per_class.npy")
confusion_matrix = np.load("./testResult/confusionMatrix.npy")
# 绘制柱状图
plt.figure(figsize=(10, 5))
plt.bar(categories, accuracy, color='skyblue')

# 显示准确率数值
for i in range(len(categories)):
    plt.text(i, accuracy[i], f'{accuracy[i]*100:.2f}%', ha='center', va='bottom')

# 添加标题和标签
plt.xlabel('Category')
plt.ylabel('Accuracy')
plt.title('Class Recognition Accuracy on Test Set')

# 显示柱状图
plt.show()

# 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=categories, yticklabels=categories)

# 添加标题和标签
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')

# 显示热力图
plt.show()

```

````python
# 绘制柱状图
plt.figure(figsize=(10, 5))
plt.bar(categories, accuracy, color='skyblue')

# 显示准确率数值
for i in range(len(categories)):
    plt.text(i, accuracy[i], f'{accuracy[i]*100:.2f}%', ha='center', va='bottom')

# 添加标题和标签
plt.xlabel('Class')
plt.ylabel('Accuracy')
plt.title('Self Model:Class Recognition Accuracy on Test Set')

# 显示柱状图
plt.show()

# 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=categories, yticklabels=categories)

# 添加标题和标签
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')

# 显示热力图
plt.show()
````

````python
# 不同模型同一张图
# 设置图的布局
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 绘制柱状图
for i, (accuracy, model_name) in enumerate(zip([accuracy, accuracy_conal, accuracy_xiaorong], model_names)):
    ax = axes[0, i]
    ax.bar(categories, accuracy, color='skyblue')
    ax.set_xticks(np.arange(len(categories)))
    ax.set_xticklabels(categories, rotation=45)
    ax.set_title(f'{model_name} Accuracy')
    ax.set_ylim(0, 1)  # 设置y轴范围

    # 显示准确率数值
    for j in range(len(categories)):
        ax.text(j, accuracy[j], f'{accuracy[j]*100:.2f}%', ha='center', va='bottom')

    # 添加标签
    ax.set_xlabel('Class')
    ax.set_ylabel('Accuracy')

# 绘制热力图
for i, confusion_matrix in enumerate([confusion_matrix, confusion_matrix_conal, confusion_matrix_xiaorong]):
    ax = axes[1, i]
    sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=categories, yticklabels=categories, ax=ax)
    ax.set_xticks(np.arange(len(categories)))
    ax.set_xticklabels(categories, rotation=45)
    ax.set_title(f'{model_names[i]} Confusion Matrix ')
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')

# 调整子图间距
plt.tight_layout()

# 显示图形
plt.show()
````
